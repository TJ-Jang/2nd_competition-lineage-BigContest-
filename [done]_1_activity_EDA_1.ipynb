```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import os
print(os.listdir('./Data/'))
```

```python
tr_act = pd.read_csv('./Data/train_activity.csv')
t1_act = pd.read_csv('./Data/test1_activity.csv')
t2_act = pd.read_csv('./Data/test2_activity.csv')
```

```python
print(tr_act.shape)
tr_act.head()
```

```python
print(t1_act.shape)
t1_act.head()
```

```python
print(t2_act.shape)
t2_act.head()
```

```python
# columns Data extraction
act_columns = tr_act.columns
```

```python
act_columns
```

```python
act_columns = act_columns.drop(['day', 'acc_id', 'char_id', 'server'])
```

```python
for col in act_columns:
    for acc in tr_act.acc_id.unique():
        tr_act.loc[(tr_act.acc_id == acc), col] = tr_act.loc[(tr_act.acc_id == acc), col].sum()
    print('tr', col, 'complete')

for col in act_columns:
    for acc in t1_act.acc_id.unique():
        t1_act.loc[(t1_act.acc_id == acc), col] = t1_act.loc[(t1_act.acc_id == acc), col].sum()
    print('t1', col, 'complete')

for col in act_columns:
    for acc in t2_act.acc_id.unique():
        t2_act.loc[(t2_act.acc_id == acc), col] = t2_act.loc[(t2_act.acc_id == acc), col].sum()
    print('t2', col, 'complete')


del(act_columns)
```

```python
tr_act = tr_act.drop(['day', 'server', 'char_id'], axis = 1)
t1_act = t1_act.drop(['day', 'server', 'char_id'], axis = 1)
t2_act = t2_act.drop(['day', 'server', 'char_id'], axis = 1)
```

```python
tr_act = tr_act.drop_duplicates()
t1_act = t1_act.drop_duplicates()
t2_act = t2_act.drop_duplicates()
```

```python
tr_act.to_csv('./input/EDA(1)_train_activity.csv', index = False)
t1_act.to_csv('./input/EDA(1)_test1_activity.csv', index = False)
t2_act.to_csv('./input/EDA(1)_test2_activity.csv', index = False)
```
